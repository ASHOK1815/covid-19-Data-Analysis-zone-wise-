```python
state=[]
for x in df['State']:
    state.append(x)
Zone=[]
for y in df['Zone']:
    Zone.append(y)
```

```python
M=list(zip(state,Zone))
```

```python
Red_zone=[]
green_zone=[]
orange_zone=[]
for x in M:
    if(x[1]=='Red Zone'):
        Red_zone.append(x[0])
    if(x[1]=='Green Zone'):
        green_zone.append(x[0])
    if(x[1]=='Orange Zone'):
        orange_zone.append(x[0])
```

```python
d={}
for x in Red_zone:
    d[x]=Red_zone.count(x)
    
```

```python
Red_z=[]
Red_count=[]
for x,y in d.items():
    Red_z.append(x)
    Red_count.append(y)
    
```

```python

plt.barh(Red_z,Red_count,color="#FF0000",label="Red Zone")
plt.legend()
plt.tight_layout()
plt.ylabel("States")
plt.xlabel("Number of Zones")
plt.show()
```

```python
d={}
for x in green_zone:
    d[x]=green_zone.count(x)
```

```python
green_z=[]
green_count=[]
for x,y in d.items():
    green_z.append(x)
    green_count.append(y)
```
